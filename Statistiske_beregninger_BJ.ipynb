```python
import numpy as np
import math
import scipy 
import scipy.stats as stats
```

```python
# Sample statistics 6 deck CA vs. 1 deck BA
mean_CA = 0.416973
mean_BA = 0.419008

std_CA = 0.001392
std_BA = 0.002147

n_CA = 15
n_BA = 15

# Forskellen i gennemsnit
mean_diff = mean_CA - mean_BA

# Standard error (Welch)
SE = math.sqrt((std_CA**2) / n_CA + (std_BA**2) / n_BA)

# Welch frihedsgrader
df = ((std_CA**2 / n_CA + std_BA**2 / n_BA)**2) / (((std_CA**2 / n_CA)**2) / (n_CA - 1) + ((std_BA**2 / n_BA)**2) / (n_BA - 1))

# 95% kritisk t-værdi
t_crit = stats.t.ppf(0.975, df)

# Confidence interval
ci_lower = mean_diff - t_crit * SE
ci_upper = mean_diff + t_crit * SE
print(f"kritisk t værdi: {t_crit:.3f}")
print(f"T-statistik: {mean_diff / SE:.3f}")
print(f"Frihedsgrader (Welch): {df:.2f}")
print(f"95% konfidensinterval: [{ci_lower:.6f} ; {ci_upper:.6f}]")

# Da 0 ikke er inden for konfidensintervallet, kan vi konkludere, at der er en signifikant forskel mellem de agenter.
# Derudover kan det konkluderes, at CA agenten er bedre end BA agenten, 
# da forskellen i middelværdier er positiv, og med 95% sikkerhed kan det konkluderes at det ligger indenfor intervallet.
```
